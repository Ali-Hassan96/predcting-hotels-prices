```python
import pandas as pd
import numpy as np
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

%matplotlib inline
```

```python
df = pd.read_csv("ANYTHING1.csv")
df
```

```python
df['price_for'] = pd.to_numeric(df['price_for'], errors='coerce')
df = df.replace(np.nan, 0, regex=True)

df['room_type'] = pd.to_numeric(df['room_type'], errors='coerce')
df = df.replace(np.nan, 0, regex=True)

df['beds'] = pd.to_numeric(df['beds'], errors='coerce')
df = df.replace(np.nan, 0, regex=True)

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df = df.replace(np.nan, 0, regex=True)

df['rating_title'] = pd.to_numeric(df['rating_title'], errors='coerce')
df = df.replace(np.nan, 0, regex=True)

df['name'] = pd.to_numeric(df['name'], errors='coerce')
df = df.replace(np.nan, 0, regex=True)

df['location'] = pd.to_numeric(df['location'], errors='coerce')
df = df.replace(np.nan, 0, regex=True)

print(df)
print(df.dtypes)
```

```python
c=['price','price_for','room_type','beds','rating', 'rating_title']
df1=df[c]
```

```python
df1=pd.get_dummies(df,columns=['price_for'], drop_first=True)
```

```python
y=df['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(df,y,train_size=0.8,random_state=42)
```

```python
x_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor()
```

```python
reg.fit(x_train,y_train)
```

```python
x_axis = x_test.rating

plt.scatter(x_axis, y_test, c = 'b', alpha = 0.5, marker = '.', label = 'Real')
plt.scatter(x_axis, reg.predict(x_test), c = 'r', alpha = 0.5, marker = '.', label = 'predicted')
plt.xlabel('rating')
plt.ylabel('price')
plt.grid(color = '#D3D3D3', linestyle = 'solid')
plt.legend(loc = 'upper left')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

mse = mean_squared_error(y_test.values.ravel(), reg.predict(x_test))
print("Mean squared error (MSE): ", round(mse, 2))

r2 = r2_score(y_test.values.ravel(), reg.predict(x_test))
print("R2 Score: ", round(r2, 2))
```
